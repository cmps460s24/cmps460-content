Euclidean Distance is a measure of the 
straight-line distance between two points (ùíô<sub>ùüè</sub>, ùíö<sub>ùüè</sub>)(ùíô<sub>2</sub>, ùíö<sub>ùüè</sub>) in a two-dimensional space


```python
# Review of zip function
languages = ['Java', 'Python', 'JavaScript']
versions = [14, 3, 6]

result = zip(languages, versions)
print(list(result))

# Output: [('Java', 14), ('Python', 3), ('JavaScript', 6)]
```

```python
import math

def euclidean_distance(point1, point2):
    distance = math.sqrt(sum([(p - q) ** 2 for p, q in zip(point1, point2)]))
    return distance

# Example points
point1 = (2, 3)
point2 = (5, 7)
print(list(zip(point1, point2)))

# Calculate Euclidean distance
distance = euclidean_distance(point1, point2)
print("Euclidean distance between", point1, "and", point2, "is:", distance)
```

When dealing with instances (data objects) with multiple n features, the Euclidean distance between two instances p and q with features (ùíë<sub>ùüè</sub>, ùíë<sub>2</sub>  , ‚Ä¶ùíë<sub>n</sub>) and (ùíí<sub>ùüè</sub>, ùíí<sub>2</sub>  , ‚Ä¶ùíí<sub>n</sub>‚Äã) respectively, is given by the square root of the sum of the squared differences between corresponding feature values:


```python
import numpy as np

# Define two instances with multiple features
instance1 = np.array([2, 3, 5, 7])
instance2 = np.array([4, 1, 6, 9])

# Calculate Euclidean Distance between the two instances
euclidean_distance = np.sqrt(np.sum((instance1 - instance2) ** 2))

print("Euclidean Distance between instance1 and instance2:", euclidean_distance)
```
