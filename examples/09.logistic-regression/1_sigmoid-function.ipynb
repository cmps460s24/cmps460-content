The sigmoid function, also known as the logistic function, is a function that maps any real number into a range between 0 and 1. The sigmoid function is defined as:

$\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(suppress = True,
   formatter = {'float_kind':'{:f}'.format})

def multi2(x):
    return x / 2

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print("sigmoid(0) = ", sigmoid(0))
print("sigmoid([7, 3, 4, -9, -6]) = ", sigmoid(np.array([7, 3, 4, -9, -6])))
print("sigmoid([7, 3, 4, -9, -6]) = ", multi2(np.array([7, 3, 4, -9, -6])))

# Generate an array of x values
# The range is from -10 to 10 with a step of 0.1
x = np.linspace(-10, 10, 100)
print("x = ", x)


# Calculate the sigmoid values
y = sigmoid(x)
print("y = ", y)

# Print the values of x and y
for i in range(len(x)):
    print("x = %.4f, y = %.8f" % (x[i], y[i]))
#print("y = %.4f", y)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Threshold (0.5)')
plt.legend()
plt.show()
```



In this example, we define the sigmoid function and apply it to a range of input values. The resulting plot shows how the sigmoid function maps any input value into the range between 0 and 1.
